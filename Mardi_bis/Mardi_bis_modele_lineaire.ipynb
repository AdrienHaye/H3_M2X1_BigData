```python
#importer vos librairies 

import pandas as pd 
import numpy as np 
from sklearn import linear_model 
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#charger les données 
#price_availability.csv
#listings_final.csv
#attention aux valeurs manquantes !!

df_pa = pd.read_csv('./data/price_availability.csv', sep=";")
df_lf = pd.read_csv('./data/listings_final.csv', sep=";")
pd.set_option('display.max_columns', None)
```

```python
df_pa = df_pa.groupby("listing_id").mean()
```

```python
df = pd.merge(df_pa,df_lf,on=['listing_id'])
df
```

```python
#définir vos variables de travail X et Y

y = np.array(df.local_price)
X = np.array([df.person_capacity,df.bedrooms, df.bathrooms])
X.shape
```

```python
#utiliser la méthode split de sklearn en splitant avec un alpha=30 et un random state=42 
#afficher la shape de vos données 

X = X.reshape((999,3))
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
y_train.shape
```

```python
#créer l'objet de régression et entrainer le sur notre ensemble d'entraînement

lr = LinearRegression().fit(X_train, y_train)
lr
```
